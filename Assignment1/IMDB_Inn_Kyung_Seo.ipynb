```python
from tensorflow.keras.datasets import imdb
from tensorflow.keras import models, layers
import numpy as np
import pandas as pd
import tensorflow as tf

# ---------------------------
# 0) Reproducibility
# ---------------------------
np.random.seed(42)
tf.random.set_seed(42)

# ---------------------------
# 1) Load + Vectorize data
# ---------------------------
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10000)

def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.
    return results

x_train = vectorize_sequences(train_data)
x_test  = vectorize_sequences(test_data)

y_train = np.asarray(train_labels).astype("float32")
y_test  = np.asarray(test_labels).astype("float32")

results_summary = []

# ---------------------------
# Model 1: 1 hidden layer (16), ReLU, BCE
# ---------------------------
model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
history1 = model.fit(x_train, y_train, epochs=20, batch_size=512, validation_split=0.2, verbose=0)
results1 = model.evaluate(x_test, y_test, verbose=0)

results_summary.append({
    "Model": "1 hidden (16), ReLU, BCE",
    "Best Val Acc": max(history1.history["val_accuracy"]),
    "Test Acc": results1[1]
})

# ---------------------------
# Model 2: 3 hidden layers (16,16,16), ReLU, BCE
# ---------------------------
model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
history2 = model.fit(x_train, y_train, epochs=20, batch_size=512, validation_split=0.2, verbose=0)
results2 = model.evaluate(x_test, y_test, verbose=0)

results_summary.append({
    "Model": "3 hidden (16,16,16), ReLU, BCE",
    "Best Val Acc": max(history2.history["val_accuracy"]),
    "Test Acc": results2[1]
})

# ---------------------------
# Model 3: 2 hidden layers (32,32), ReLU, BCE
# ---------------------------
model = models.Sequential()
model.add(layers.Dense(32, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
history3 = model.fit(x_train, y_train, epochs=20, batch_size=512, validation_split=0.2, verbose=0)
results3 = model.evaluate(x_test, y_test, verbose=0)

results_summary.append({
    "Model": "2 hidden (32,32), ReLU, BCE",
    "Best Val Acc": max(history3.history["val_accuracy"]),
    "Test Acc": results3[1]
})

# ---------------------------
# Model 4: 2 hidden layers (64,64), ReLU, BCE
# ---------------------------
model = models.Sequential()
model.add(layers.Dense(64, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
history4 = model.fit(x_train, y_train, epochs=20, batch_size=512, validation_split=0.2, verbose=0)
results4 = model.evaluate(x_test, y_test, verbose=0)

results_summary.append({
    "Model": "2 hidden (64,64), ReLU, BCE",
    "Best Val Acc": max(history4.history["val_accuracy"]),
    "Test Acc": results4[1]
})

# ---------------------------
# Model 5: 2 hidden layers (32,32), tanh, BCE
# ---------------------------
model = models.Sequential()
model.add(layers.Dense(32, activation='tanh', input_shape=(10000,)))
model.add(layers.Dense(32, activation='tanh'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
history5 = model.fit(x_train, y_train, epochs=20, batch_size=512, validation_split=0.2, verbose=0)
results5 = model.evaluate(x_test, y_test, verbose=0)

results_summary.append({
    "Model": "2 hidden (32,32), tanh, BCE",
    "Best Val Acc": max(history5.history["val_accuracy"]),
    "Test Acc": results5[1]
})

# ---------------------------
# Model 6: 2 hidden layers (32,32), ReLU, MSE
# ---------------------------
model = models.Sequential()
model.add(layers.Dense(32, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop', loss='mse', metrics=['accuracy'])
history6 = model.fit(x_train, y_train, epochs=20, batch_size=512, validation_split=0.2, verbose=0)
results6 = model.evaluate(x_test, y_test, verbose=0)

results_summary.append({
    "Model": "2 hidden (32,32), ReLU, MSE",
    "Best Val Acc": max(history6.history["val_accuracy"]),
    "Test Acc": results6[1]
})

# ---------------------------
# Model 7: 2 hidden layers (32,32), ReLU + Dropout, BCE
# ---------------------------
model = models.Sequential()
model.add(layers.Dense(32, activation='relu', input_shape=(10000,)))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
history7 = model.fit(x_train, y_train, epochs=20, batch_size=512, validation_split=0.2, verbose=0)
results7 = model.evaluate(x_test, y_test, verbose=0)

results_summary.append({
    "Model": "2 hidden (32,32), ReLU, BCE + Dropout",
    "Best Val Acc": max(history7.history["val_accuracy"]),
    "Test Acc": results7[1]
})

# ---------------------------
# Summary Table
# ---------------------------
df = pd.DataFrame(results_summary)
df = df.sort_values(by="Best Val Acc", ascending=False)
print(df)

df.to_csv("imdb_results_summary.csv", index=False)
print("\nSaved: imdb_results_summary.csv")
```
